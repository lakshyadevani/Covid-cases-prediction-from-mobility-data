##K-means Clustering Implementation

```python
data = df[['retail_and_recreation',	'grocery_and_pharmacy',	'residential',	'transit_stations',	'parks',	'workplaces']].values
y_data = df[['new_cases']].values
#data.shape = (471, 6)
```

```python
len(data)
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
k = 5
iter = 100
```

```python
random_indices = np.random.choice(data.shape[0], size = k)
centroids = data[random_indices, :]    #These centroids are random initialisations
```

```python
def update_assignments(data, centroids):
    assignments = []   #An array that tells which cluster a particular data point is attached to.
    for i in data:
        assignments.append(np.argmin(np.sum((i - centroids)**2, axis=1)))
    return assignments

def update_centroids(data, k, assignments):
    new_centroids = []    #Updating centroids with every new iteration.
    for c in range(k):
        new_centroids.append(np.mean([data[x] for x in range(len(data)) if assignments[x] == c], axis=0))
    return new_centroids
```

```python
for i in range(iter):
  assignments = update_assignments(data, centroids)
  centroids = update_centroids(data, k, assignments)
  centroids = np.array(centroids)
```

```python
#assignments
cluster_dict = {}
y_cluster_dict = {}

for i in range(k):
  cluster_dict[i]=[]
  y_cluster_dict[i]=[]
  for j in range(len(assignments)):
    if(i == assignments[j]):
      cluster_dict[i].append(data[j])
      y_cluster_dict[i].append(y_data[j])
#y_cluster_dict


```

##Decision Tree Regression - Implementation

```python
class Node():
    def __init__(self, feature_index=None, threshold=None, left=None, right=None, var_red=None, value=None):
        ''' constructor ''' 
        
        # for decision node
        self.feature_index = feature_index
        self.threshold = threshold
        self.left = left
        self.right = right
        self.var_red = var_red
        
        # for leaf node
        self.value = value
```

```python
def train_test_split_code(data, y_data, test_size, randomState):
  training_data = data.sample(frac=1-test_size, random_state=randomState)
  testing_data = data.drop((training_data.index))

  training_y_data = y_data.drop((testing_data.index))
  testing_y_data = y_data.drop((training_y_data.index))
  
  return(training_data, testing_data, training_y_data, testing_y_data)
```

```python
import math
def root_mean_squared_error(ls1,ls2):
  
  y_actual = ls1
  y_predicted = ls2

  MSE = np.square(np.subtract(y_actual,y_predicted)).mean() 
  RMSE = math.sqrt(MSE)
  return(RMSE)
```

```python
class DecisionTreeRegressor():
    def __init__(self, min_samples_split=2, max_depth=2):        
        # initialize the root of the tree 
        self.root = None
        
        # stopping conditions
        self.min_samples_split = min_samples_split
        self.max_depth = max_depth

    def build_tree(self, dataset, curr_depth=0):   #Recursive function to build a tree
      X, Y = dataset[:,:-1], dataset[:,-1]   #Splitting the dataset as features and labels

      num_samples, num_features = np.shape(X)    

      best_split = {}     #The best_split function of this class returns a dictionary, so initialising the library

      if num_samples >= self.min_samples_split and curr_depth <= self.max_depth:    #Stopping Conditions

        best_split = self.get_best_split(dataset, num_samples, num_features)    #Finding the best split
        
        if best_split["var_red"]>0:    #Checking if the information gain is positive
            left_subtree = self.build_tree(best_split["dataset_left"], curr_depth+1)
            right_subtree = self.build_tree(best_split["dataset_right"], curr_depth+1)
            return Node(best_split["feature_index"], best_split["threshold"], 
                        left_subtree, right_subtree, best_split["var_red"])
              
      leaf_value = self.calculate_leaf_value(Y)    #After the tree is recursively built, calculating and returning the leaf nodes.
      return Node(value=leaf_value) 

    def get_best_split(self, dataset, num_samples, num_features):
      best_split = {}   #dictionary to store the best split
      max_var_red = -float("inf")   #Unbounded Lower Value for Comparision
      
      means=np.mean(dataset,axis=0)
      for feature_index in range(num_features):
          #Selection of a particular feature of all the samples.
          threshold=means[feature_index]#
          dataset_left, dataset_right = self.split(dataset, feature_index, threshold)   #get current split

          if len(dataset_left) > 0 and len(dataset_right) > 0:    #check if childs are not null
            y = dataset[:, -1]
            left_y = dataset_left[:, -1]
            right_y = dataset_right[:, -1]
            curr_var_red = self.variance_reduction(y, left_y, right_y)    #Calculating the Information Gain

            if curr_var_red > max_var_red:     #Check whether this is feature has the best split or not!
              best_split["feature_index"] = feature_index
              best_split["threshold"] = threshold
              best_split["dataset_left"] = dataset_left
              best_split["dataset_right"] = dataset_right
              best_split["var_red"] = curr_var_red

              max_var_red = curr_var_red   #If this is the best split then update the maximum Information Gain.

      return best_split

    def split(self, dataset, feature_index, threshold):
      dataset_left = np.array([row for row in dataset if row[feature_index] <= threshold])   #All the rows of the dataset for which a particular feature's value is less than the threshold.
      dataset_right = np.array([row for row in dataset if row[feature_index]>threshold])     #All the rows of the dataset for which a particular feature's value is more than the threshold.

      return dataset_left, dataset_right

    def variance_reduction(self, parent, l_child, r_child):   #Calculating the weighted 
      weight_l = len(l_child) / len(parent)
      weight_r = len(r_child) / len(parent)
      reduction = np.std(parent) - (weight_l * np.std(l_child) + weight_r * np.std(r_child))

      return reduction

    def calculate_leaf_value(self, Y):
      val = np.mean(Y)
      return val

    def print_tree(self, tree=None, indent=" "):
      if not tree:
        tree = self.root
      if tree.value is not None:
        print(tree.value)
      else:
        print("X_"+str(tree.feature_index), "<=", tree.threshold, "?", tree.var_red)
        print("%sleft:" % (indent), end="")
        self.print_tree(tree.left, indent + indent)
        print("%sright:" % (indent), end="")
        self.print_tree(tree.right, indent + indent)

    def fit(self, X, Y):
      dataset = np.concatenate((X, Y), axis=1)
      self.root = self.build_tree(dataset)

    def make_prediction(self, x, tree):
      if tree.value != None: 
        return tree.value
      feature_val = x[tree.feature_index]
      
      if feature_val<=tree.threshold:
          return self.make_prediction(x, tree.left)
      else:
          return self.make_prediction(x, tree.right)
    
    def predict(self, X):
      preditions = [self.make_prediction(x, self.root) for x in X]
      return preditions
```
